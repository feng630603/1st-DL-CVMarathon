```python
"""
瀏覽器開發者工具介紹
知道瀏覽器的開發者工具
能夠使用瀏覽器的開發者工具觀察資料
作業目標
利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。

1. https://www.zhihu.com/explore
"""
```

```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

```python
import requests
url = 'http://www.zhihu.com'

headers = {https://www.aicoin.cn/embed/ws/whoami?publicKey=bf35dad6-1d25-48dc-b460-41ab648659e6}
   

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
```
